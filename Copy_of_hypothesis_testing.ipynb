## Question 1: What is a null hypothesis (H₀) and why is it important in hypothesis testing?

### Answer:
A null hypothesis (H₀) is a statement which assumes that there is no change, no effect, or no difference in the population. It acts as the starting point for any hypothesis test.  
It is important because the whole test is done to check whether we have enough evidence to reject this assumption or not.


## Question 2: What does the significance level (α) represent in hypothesis testing?

### Answer:
The significance level α tells us the probability of making a Type I error, which means rejecting the null hypothesis when it is actually true.  
Common values are 0.05 and 0.01, meaning we accept a 5% or 1% chance of such an error.


## Question 3: Differentiate between Type I and Type II errors.

### Answer:
Type I Error: Rejecting the null hypothesis when it is actually true (false positive).  
Type II Error: Not rejecting the null hypothesis when it is actually false (false negative).


## Question 4: Explain the difference between a one-tailed and two-tailed test. Give an example.

### Answer:
A one-tailed test checks for an effect in only one direction (either greater than or less than some value).  
Example: H₁: μ > 10  

A two-tailed test checks for an effect in both directions.  
Example: H₁: μ ≠ 10


## Question 5: A company claims that the average time to resolve a complaint is 10 minutes.
A sample of 9 complaints gives mean = 12 and SD = 3. Test the claim at α = 0.05.

### Answer:
Here n is small and population SD is not given, so we use the t-test.

H₀: μ = 10  
H₁: μ ≠ 10  

t = (12 − 10) / (3/√9)  
t = 2  

df = 8  
Critical value (two-tailed, α=0.05) = ±2.306  

Since 2 < 2.306, we do not reject H₀.

Conclusion: There is not enough evidence to say that the average time is different from 10 minutes.


## Question 6: When should you use a Z-test instead of a t-test?

### Answer:
A Z-test is used when the population standard deviation is known or when the sample size is large (30 or more).  
A t-test is used when the population standard deviation is unknown and the sample size is small.

So:  
Z-test → known SD or large sample  
t-test → unknown SD and small sample


## Question 7: Productivity of 6 employees before and after training was recorded. Test if training improved performance at α = 0.05.

### Answer:
This is a paired t-test.

Differences (After − Before): 5, 5, 1, 3, 1, 3  
Mean difference = 3  
SD of differences ≈ 1.63  
n = 6  

H₀: μd = 0  
H₁: μd > 0  

t ≈ 4.50  
Critical value (df=5, one-tailed) = 2.015  

Since 4.50 > 2.015, we reject H₀.

Conclusion: The training program improved productivity.


## Question 8: Test if product preference is independent of gender at α = 0.05.

### Answer:
We use the chi-square test.

Observed:  
Male: A=30, B=20  
Female: A=40, B=10  

Expected values:  
Male-A=35, Male-B=15  
Female-A=35, Female-B=15  

χ² = Σ (O−E)²/E = 4.76  

df = 1  
Critical value = 3.84  

Since 4.76 > 3.84, we reject H₀.

Conclusion: Product preference depends on gender (not independent).
